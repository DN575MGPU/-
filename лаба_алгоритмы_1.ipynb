```python
class CallCenter:
    """
    Класс для симуляции работы колл-центра
    Использует очередь (принцип FIFO)
    """

    def __init__(self):  # Вызывается автоматически при создании объекта
        """Инициализация пустой очереди звонков"""
        self.queue = []  # Используем список как очередь

    def add_call(self, client_name):
        """
        Добавить новый звонок в очередь
        (постановка в конец очереди)
        """
        self.queue.append(client_name)
        print(f"  Звонок от {client_name} поставлен в очередь")
        print(f"  Всего в очереди: {len(self.queue)} звонков")
        print()

    def process_call(self):
        """
        Обработать следующий звонок
        (удаление из начала очереди)
        """
        if self.is_empty():  # Проверка пуста ли очередь
            print("В очереди нет звонков!")
            return None      # если очередь пуста, возвращаем "ничего"

        # Удаляем и возвращаем первый элемент очереди
        client = self.queue.pop(0)
        print(f"  Обрабатывается звонок от {client}")
        print(f"  Осталось в очереди: {len(self.queue)} звонков")
        print()
        return client        # если звонок есть, возвращаем клиента

    def show_queue(self):
        """Показать текущую очередь звонков"""
        if self.is_empty():
            print(" Очередь пуста")
            print()
        else:
            print(" Текущая очередь звонков:")
            for i, client in enumerate(self.queue, 1):  # функция enumerate, которая нумерует элементы
                print(f"  {i}. {client}")
            print(f"  Всего: {len(self.queue)} звонков")  # код перебирает всех клиентов в self.queue и выводит нумерованный список, начиная с 1
            print()

    def is_empty(self):
        """Проверить, пуста ли очередь"""
        return len(self.queue) == 0  # Возвращает True, если очередь пуста (длина = 0), если есть звонки - False

    def get_queue_length(self):
        """Получить количество звонков в очереди"""
        return len(self.queue)      # Возвращает количество звонков в очереди

# ========== СИМУЛЯЦИЯ РАБОТЫ КОЛЛ-ЦЕНТРА ==========

def simulate_call_center():
    """Функция для симуляции работы колл-центра"""

    # Создаем колл-центр
    call_center = CallCenter()

    print()
    print(" СИМУЛЯЦИЯ РАБОТЫ КОЛЛ-ЦЕНТРА")
    print()

    # 1. Добавляем звонки в очередь
    print("[1] ПОСТАНОВКА ЗВОНКОВ В ОЧЕРЕДЬ:")
    print()

    # Запрашиваем у пользователя количество клиентов
    while True:                    # Бесконечный цикл
        try:                       # Пытаемся выполнить "опасный" код
            num_clients = int(input("Сколько клиентов хотите добавить? "))
            if num_clients > 0:   # Проверяем, что число положительное
                break             # выход из цикла
            else:
                print("Пожалуйста, введите число больше 0")
        except ValueError:
            print("Пожалуйста, введите целое число")

    print()

    # Вводим данные каждого клиента
    for i in range(num_clients):
        print(f"Клиент {i+1}:")
        client_name = input("  Введите имя и фамилию: ").strip() # удаление пробелов по краям строки

        # Проверка, что имя не пустое
        if client_name:                        # Если имя НЕ пустое (True)
            call_center.add_call(client_name)   # добавляем в колл центр
        else:
            print("  Имя не может быть пустым! Попробуйте снова.")
            # Даем еще одну попытку
            client_name = input("  Введите имя и фамилию: ").strip()
            if client_name:
                call_center.add_call(client_name)  #если имя есть - добавляем
            else:
                call_center.add_call(f"Клиент {i+1}") # Если опять пусто - используем имя по умолчанию (Клиент 1, 2...)

    # 2. Показываем текущую очередь
    print()
    print("[2] ТЕКУЩАЯ ОЧЕРЕДЬ ЗВОНКОВ:")
    print()
    call_center.show_queue()


    # 3. Спрашиваем, хочет ли пользователь обработать звонки
    print()
    print("[3] ОБРАБОТКА ЗВОНКОВ:")
    print()

    process_choice = input("Хотите обработать все звонки? (да/нет): ").strip().lower()

    if process_choice in ['да', 'yes']:
        # Обрабатываем все звонки
        print(" Начинаем обработку звонков...")

        while not call_center.is_empty():  # ПОКА очередь НЕ пуста
            call_center.process_call()     # обрабатываем один звонок


        # 4. Проверяем, что очередь пуста
        print()
        print("[4] СОСТОЯНИЕ ОЧЕРЕДИ ПОСЛЕ ОБРАБОТКИ:")
        print()
        call_center.show_queue()
    else:
        print(" Обработка звонков отменена.")
        print(f"В очереди осталось {call_center.get_queue_length()} звонков")

# Запускаем симуляцию
simulate_call_center()
```
